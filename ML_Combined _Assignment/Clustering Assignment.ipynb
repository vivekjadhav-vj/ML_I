In this assignment students have to compress racoon grey scale image into
5 clusters. In the end, visualize both raw and compressed image and look
for quality difference.
The raw image is available in spicy.misc package with the name face.

```python
import numpy as np
from sklearn import datasets
from scipy import misc
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
face= misc.face(gray=True)
```

```python
n_clusters= 5
np.random.seed(0)

x= face.reshape((-1,1))
k_means = KMeans(n_clusters=n_clusters, n_init=4)
k_means.fit(x)
values= k_means.cluster_centers_.squeeze()
lables= k_means.labels_

#create an array from lables and values
face_compressed =np.choose(lables, values)
face_compressed.shape = face.shape

vmin = face.min()
vmax = face. max()
```

```python
#original face
plt.figure(1)
plt.imshow(face, cmap=plt.cm.gray, vmin=vmin, vmax=300)
```

```python
# Compressed face
plt.figure(2)
plt.imshow(face_compressed, cmap=plt.cm.gray, vmin=vmin, vmax=vmax)
```
